```python

import pandas as pd
import string
import random

# Load the Excel file
try:
  df = pd.read_excel('/content/NLP_dataset.xlsx')
except FileNotFoundError:
  print("Error: File not found at /content/NLP_dataset.xlsx")
  exit()
# Number of new rows to generate
num_new_rows = 10  # Change this to the desired number

# Generate new rows with random data
new_rows = []
for _ in range(num_new_rows):
    new_row = {}
    for col in df.columns:
      if pd.api.types.is_string_dtype(df[col]):
        # Generate random string data (adjust as needed for your data)
          new_row[col] = ''.join(random.choice(string.ascii_letters) for _ in range(10))
      elif pd.api.types.is_numeric_dtype(df[col]):
        # Generate random numeric data (adjust range as needed)
          new_row[col] = random.randint(0, 100)
      else:
          # Handle other data types (e.g., datetime) as required
          new_row[col] = df[col].iloc[0] # Or some other default value
    new_rows.append(new_row)

# Create a new DataFrame from the generated rows
new_df = pd.DataFrame(new_rows, columns=df.columns)

# Concatenate the original DataFrame with the new rows
df = pd.concat([df, new_df], ignore_index=True)

print(f"The dataset now has {len(df)} rows.")

# Save to a new Excel file
df.to_excel('cleaned_NLP_dataset.xlsx', index=False)
```

```python
import os
import pandas as pd
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

def spacy_sent_tokenize(text):
    return [sent.text.strip() for sent in nlp(text).sents]

def is_formal(text):
    text = text.lower()
    if any(bad in text for bad in ["lol", "haha", "wassup", "bro", "gonna", "ain't", "yo"]):
        return False
    if len(text.split()) < 20:
        return False
    if any(text.startswith(greet) for greet in ["dear", "hello", "hi", "good morning"]):
        return True
    if any(phrase in text for phrase in ["please find", "i would like", "kindly", "i am writing", "i look forward"]):
        return True
    return False

def extract_email_body(email_text):
    parts = email_text.split("\n\n", 1)
    return parts[1] if len(parts) == 2 else email_text

formal_set = set()
count = 0
TARGET_COUNT = 2000
MAILDIR_PATH = "/content/maildir"
OUTPUT_PATH = f"/content/formal_enron_{TARGET_COUNT}.csv"

if os.path.exists(MAILDIR_PATH):
    for root, _, files in os.walk(MAILDIR_PATH):
        for filename in files:
            filepath = os.path.join(root, filename)
            try:
                with open(filepath, "r", encoding="latin-1") as f:
                    raw_email = f.read()
                    body = extract_email_body(raw_email)
                    sentences = spacy_sent_tokenize(body)
                    for sent in sentences:
                        if is_formal(sent):
                            formal_set.add(sent)
                            count += 1
                            if count % 1000 == 0:
                                print(f"‚úÖ {count} formal sentences collected...")
                            if count >= TARGET_COUNT:
                                break
            except Exception as e:
                print(f"‚ùå Skipped {filepath}: {e}")
                continue
        if count >= TARGET_COUNT:
            break
else:
    print(f"‚ùå Directory not found: {MAILDIR_PATH}")

df = pd.DataFrame(list(formal_set), columns=["Formal_Email_Sentence"])
df.to_csv(OUTPUT_PATH, index=False)
print(f"\nüéâ Done! Saved {len(df)} formal sentences to {OUTPUT_PATH}")

```

```python
import os
import pandas as pd
import spacy

# Load spaCy English model
try:
    nlp = spacy.load("en_core_web_sm")
except:
    import subprocess
    subprocess.run(["python", "-m", "spacy", "download", "en_core_web_sm"])
    nlp = spacy.load("en_core_web_sm")

def spacy_sent_tokenize(text):
    return [sent.text.strip() for sent in nlp(text).sents]

def is_formal(text):
    text_lower = text.lower()
    word_count = len(text.split())

    if word_count < 6 or word_count > 24:
        return False

    informal_words = ["lol", "haha", "wassup", "bro", "gonna", "ain't", "yo"]
    if any(word in text_lower for word in informal_words):
        return False

    if any(text_lower.startswith(greet) for greet in ["dear", "hello", "hi", "good morning"]):
        return True
    if any(phrase in text_lower for phrase in ["please find", "i would like", "kindly", "i am writing", "i look forward"]):
        return True

    return False

def extract_email_body(text):
    parts = text.split("\n\n", 1)
    return parts[1] if len(parts) == 2 else text

# Initialize
formal_sentences = set()
count = 0
TARGET = 5000
MAILDIR_PATH = "/content/maildir"
OUTPUT_PATH = "/content/formal_enron_spacy_5000.csv"

if os.path.exists(MAILDIR_PATH):
    for root, _, files in os.walk(MAILDIR_PATH):
        for filename in files:
            filepath = os.path.join(root, filename)
            try:
                with open(filepath, "r", encoding="latin-1") as f:
                    email = f.read()
                    body = extract_email_body(email)
                    sentences = spacy_sent_tokenize(body)

                    for sent in sentences:
                        if is_formal(sent):
                            formal_sentences.add(sent)
                            count += 1
                            if count % 1000 == 0:
                                print(f"‚úÖ Collected {len(formal_sentences)} formal sentences...")
                            if len(formal_sentences) >= TARGET:
                                break
            except Exception as e:
                print(f"‚ùå Error with {filepath}: {e}")
                continue
        if len(formal_sentences) >= TARGET:
            break
else:
    print(f"‚ùå Directory not found: {MAILDIR_PATH}")

# Save to CSV
df = pd.DataFrame(list(formal_sentences)[:TARGET], columns=["Formal_Email_Sentence"])
df.to_csv(OUTPUT_PATH, index=False)
print(f"\nüéâ Done! Saved {len(df)} formal sentences to {OUTPUT_PATH}")
```

```python
# prompt: /content/formal_enron_spacy_5000.csv Clean the above dataset, remove duplicate, lower case, uneccesary symbols and show the shape and then save cleaned

import pandas as pd
import string

# Load the CSV file
df = pd.read_csv('/content/formal_enron_spacy_5000.csv')

# Display original shape
print("Original shape:", df.shape)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Lowercase the text
df['Formal_Email_Sentence'] = df['Formal_Email_Sentence'].str.lower()

# Remove punctuation
def remove_punctuation(text):
    if isinstance(text, str):
        translator = str.maketrans('', '', string.punctuation)
        return text.translate(translator)
    else:
        return text

df['Formal_Email_Sentence'] = df['Formal_Email_Sentence'].apply(remove_punctuation)

# Display cleaned shape
print("Cleaned shape:", df.shape)

# Save the cleaned DataFrame to a new CSV file
df.to_csv('/content/cleaned_formal_enron_spacy_5000.csv', index=False)
```
